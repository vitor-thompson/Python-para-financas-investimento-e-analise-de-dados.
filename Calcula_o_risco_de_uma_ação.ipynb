Calcula o risco de uma ação

```
# Instala a biblioteca yfinance silenciosamente, caso ainda não esteja instalada.
!pip install -q yfinance

# Importa a biblioteca numpy, essencial para operações numéricas e cálculos matemáticos.
import numpy as np
# Importa o módulo 'data' de pandas_datareader como 'wb', geralmente usado para buscar dados de fontes como o Yahoo Finance.
from pandas_datareader import data as wb
# Importa a biblioteca yfinance para baixar dados financeiros diretamente do Yahoo Finance.
import yfinance as yf
# Importa a biblioteca matplotlib.pyplot para visualização de dados, criando gráficos e plots.
import matplotlib.pyplot as plt
# Importa a biblioteca pandas, fundamental para manipulação e análise de dados em formato de tabelas (DataFrames).
import pandas as pd
```

```
# Define uma lista de tickers (símbolos de ações) das empresas brasileiras a serem analisadas.
# A lista inclui Itaú Unibanco (ITUB4), Itaúsa (ITSA4), Taesa (TAEE11), Sabesp (SBSP3), Vale (VALE3),
# Petrobras (PETR4), Raia Drogasil (RADL3), Weg (WEGE3), BB Seguridade (BBSE3) e Rumo (RAIL3).
tickers = ['ITUB4.SA','ITSA4.SA','TAEE11.SA','SBSP3.SA','VALE3.SA','PETR4.SA', 'RADL3.SA','WEGE3.SA','BBSE3.SA','RAIL3.SA']
# Define o número de dias úteis em um ano comercial, comumente usado em cálculos financeiros.
dias_uteis = 252
# Inicializa um DataFrame vazio do pandas que será usado para armazenar os dados de fechamento das ações.
meus_dados = pd.DataFrame()

# Itera sobre a lista de tickers para baixar os dados históricos de cada ação.
# Para cada ticker 't', a função yf.download() baixa os dados de fechamento ('Close') a partir de '2000-1-1'.
# Os dados de fechamento são então adicionados como uma nova coluna ao DataFrame 'meus_dados'.
for t in tickers:
    meus_dados[t] = yf.download(t, start='2000-1-1')['Close']
```

```
# Calcula os retornos logarítmicos diários das ações.
# Retornos logarítmicos são frequentemente usados em finanças por suas propriedades matemáticas (aditividade).
# A fórmula é log(Pt / Pt-1), onde Pt é o preço atual e Pt-1 é o preço anterior.
retornos_log = np.log(meus_dados / meus_dados.shift(1))
```

```
# Exibe o DataFrame contendo os retornos logarítmicos diários de cada ação.
# Isso permite uma visualização rápida dos dados calculados.
retornos_log
```

```
# Calcula o risco anual (volatilidade) de cada ativo usando o desvio padrão dos retornos logarítmicos.
# O desvio padrão diário é anualizado multiplicando-o pela raiz quadrada do número de dias úteis (250 ou 252).
risco_ativo = retornos_log.std() * 250 ** 0.5
```

```
# Calcula o retorno médio anual de cada ativo.
# A média dos retornos logarítmicos diários é multiplicada pelo número de dias úteis no ano (dias_uteis).
retornos_log.mean() * dias_uteis
```

```
# Exibe o risco anual (volatilidade) calculado para cada ativo no portfólio.
# Este valor representa a flutuação esperada no preço do ativo ao longo de um ano.
risco_ativo
```
